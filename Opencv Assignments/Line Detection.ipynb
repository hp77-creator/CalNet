```python
import cv2 
import numpy as np
```

```python
image = cv2.imread(r'C:\Users\Tanya srivastava\Downloads\sudoko.jpeg')
cv2.imshow('lines',image)
cv2.waitKey(0)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,100,170,apertureSize = 3)

lines = cv2.HoughLines(edges,1,np.pi/180,220)

for line in lines:
    rho,theta = line[0]
    a=np.cos(theta)
    b=np.sin(theta)
    x0=a*rho
    y0=b*rho
    x1=int(x0+1000*(-b))
    y1=int(y0+1000*(a))
    x2=int(x0-1000*(-b))
    y2=int(y0-1000*(a))
    cv2.line(image,(x1,y1),(x2,y2),(0,0,255),2)
    
cv2.imshow('hough lines',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
